```python
def sum_of_squares(stop_val):
    sum_ = 0
    for i in range(stop_val):
        sum_ += i**2
    return sum_
```

```python
sum_of_squares(5)
```

```python
sum_5 = sum_of_squares(5)
```

```python
def print_squares_with_startval(stop_val, startval=0):
    """It prints the squares of numbers starting from start value(inclusive) to stop val(exclusive)"""
    for i in range(startval, stop_val):
        print(i, i**2)
```

```python
print_squares_with_startval(5,2)
```

# Classes

```python
class MobilePhone:
    """This ia sample class to see how classes work"""
    def __init__(self, name, is_android=False, screen_size=4.3):
        self.name = name
        self.is_android = is_android
        self.screen_size = screen_size
        self.rating = -1
        
    def has_rating(self):
        return self.rating > -1
```

```python
new_phone = MobilePhone('iphone 5s')  #new_phone gets passed the first instance value to self
```

```python
type(new_phone)
```

```python
print(new_phone.name, new_phone.is_android, new_phone.screen_size)
```

```python
new_phone.screen_size = 4
```

```python
print(new_phone.name, new_phone.is_android, new_phone.screen_size)
```

```python
new_phone.has_rating()
```

```python
new_phone.rating = 5
```

```python
new_phone.has_rating()
```

```python
MobilePhone.__doc__
```

```python
class iPhone(MobilePhone):
    def __init__(self, name):
        MobilePhone.__init__(self, name, False, 4)
        
    def __str__(self):
        return self.name + " " + str(self.is_android) + " " + str(self.screen_size)
```

```python
new_iphone = iPhone("iPhone 5s")
```

```python
new_iphone.is_android
```

```python
print(new_iphone)
```
